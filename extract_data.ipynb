```python
def game_date(game_id):
    return  mlb.get_game_box_score(game_id).info[-1].label
```

```python
mlb.get_scheduled_games_by_date("2025-07-30")
```

```python
def is_day_game(game_id):
    return mlb.get_game(game_id).gamedata.datetime.daynight == "day"
```

```python
def home_team_won(game_id):
    game = mlb.get_game_box_score(game_id)
    return game.teams.home.teamstats["batting"]['runs'] > game.teams.away.teamstats["batting"]['runs']
```

```python
def get_game_data(game_id, batting_param, pitching_params):

    box_score = mlb.get_game_box_score(game_id)

    home_abb = box_score.teams.home.team.abbreviation
    away_abb = box_score.teams.away.team.abbreviation

    home_team_batting = box_score.teams.home.teamstats["batting"]
    away_team_batting = box_score.teams.away.teamstats["batting"]

    home_team_pitching = box_score.teams.home.teamstats["pitching"]
    away_team_pitching = box_score.teams.away.teamstats["pitching"]

    data = {}
    
    data['Date'] = game_date(game_id)

    data["Home Team"] = home_abb
    data["Away Team"] = away_abb

    for param in batting_param:
        data[f"Difference {param}"] = float(home_team_batting.get(param))- float(away_team_batting.get(param))

    for param in pitching_params:
        data[f"Difference {param}"] = float(home_team_pitching.get(param)) - float(away_team_pitching.get(param))

    data['Home Team Won'] = home_team_won(game_id)

    return data
```

```python
def get_all_game_dates(start_date, end_date):
    schedule = mlb.get_schedule(start_date=start_date, end_date=end_date)
    dates = []
    for date in schedule.dates:
        dates.append(date.date)
    return dates
```

```python
def get_all_game_ids(start_date, end_date):
    lst = []
    dates = get_all_game_dates(start_date, end_date)
    for date in dates:
        games = mlb.get_scheduled_games_by_date(date)
        for game in games:
            lst.append(game.gamepk)
    return lst
```
