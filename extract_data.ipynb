```python
def game_date(game_id):
    return  mlb.get_game_box_score(game_id).info[-1].label
```

```python
def is_day_game(game_id):
    return mlb.get_game(game_id).gamedata.datetime.daynight == "day"
```

```python
def home_team_won(game_id):
    game = mlb.get_game_box_score(game_id)
    return game.teams.home.teamstats["batting"]['runs'] > game.teams.away.teamstats["batting"]['runs']
```

```python
def innings(innings_pitched_str):
    if '.' in innings_pitched_str:
        whole, fraction = innings_pitched_str.split('.')
        return int(whole) + int(fraction) / 3
    else:
        return int(innings_pitched_str)
```

```python
def pitcher_era_home_and_away(game_id):
    box = mlb.get_game_box_score(game_id = game_id)
    home_innings = 0
    away_innings = 0
    key_home = None
    key_away = None
    for key in box.teams.home.players.keys():
        if box.teams.home.players[key].stats['pitching'] != {}:
            if innings(box.teams.home.players[key].stats['pitching']['inningspitched']) > home_innings:
                home_innings = innings(box.teams.home.players[key].stats['pitching']['inningspitched'])
                key_home = key
    for key in box.teams.away.players.keys():
        if box.teams.away.players[key].stats['pitching'] != {}:
            if innings(box.teams.away.players[key].stats['pitching']['inningspitched']) > away_innings:
                away_innings = innings(box.teams.away.players[key].stats['pitching']['inningspitched'])
                key_away = key
    if key_home is not None and key_away is not None:
        home_era = np.round(box.teams.home.players[key_home].stats['pitching']['earnedruns'] / home_innings * 9, decimals=2)
        away_era = np.round(box.teams.away.players[key_away].stats['pitching']['earnedruns'] / away_innings * 9, decimals=2)
        return box.teams.home.players[key_home].person.fullname, home_era, box.teams.away.players[key_away].person.fullname, away_era
```

```python
def get_game_data(game_id, batting_param, pitching_params):

    box_score = mlb.get_game_box_score(game_id)

    home_abb = box_score.teams.home.team.abbreviation
    away_abb = box_score.teams.away.team.abbreviation

    home_team_batting = box_score.teams.home.teamstats["batting"]
    away_team_batting = box_score.teams.away.teamstats["batting"]

    home_team_pitching = box_score.teams.home.teamstats["pitching"]
    away_team_pitching = box_score.teams.away.teamstats["pitching"]

    data = {}
    
    data['Date'] = game_date(game_id)

    data["Home Team"] = home_abb
    data["Away Team"] = away_abb

    for param in batting_param:
        data[f"Home {param}"] = float(home_team_batting.get(param))
        data[f"Away {param}"] = float(away_team_batting.get(param))

    for param in pitching_params:   
        data[f"Home {param}"] = float(home_team_pitching.get(param))
        data[f"Away {param}"] = float(away_team_pitching.get(param))

    home_pitcher, home_era, away_pitcher, away_era = pitcher_era_home_and_away(game_id)

    data['Home SP'] = home_pitcher
    data['Home SP ERA'] = home_era
    data['Away SP'] = away_pitcher
    data['Away SP ERA'] = away_era

    data['Home Team Won'] = home_team_won(game_id)

    return data
```

```python
def get_all_game_dates(start_date, end_date):
    schedule = mlb.get_schedule(start_date=start_date, end_date=end_date)
    dates = []
    for date in schedule.dates:
        dates.append(date.date)
    return dates
```

```python
def get_all_game_ids(start_date, end_date):
    lst = []
    dates = get_all_game_dates(start_date, end_date)
    for date in dates:
        games = mlb.get_scheduled_games_by_date(date)
        for game in games:
            lst.append(game.gamepk)
    return lst
```

```python


# Calculate average stats for each team at home and away (using last 81 games)
def create_team_averages(df):
    # Get all unique teams
    all_teams = set(df['Home Team'].unique()) | set(df['Away Team'].unique())
    
    # Get column names for stats (excluding non-stat columns)
    home_stat_cols = [col for col in df.columns if col.startswith('Home ') and col not in ['Home Team', 'Home Team Won', 'Home SP', 'Home SP ERA']]
    away_stat_cols = [col for col in df.columns if col.startswith('Away ') and col not in ['Away Team', 'Away SP', 'Away SP ERA']]
    
    # Get unique stat names
    stat_names = [col.replace('Home ', '') for col in home_stat_cols]
    
    team_avg_data = []
    
    for team in all_teams:
        # Home games for this team
        home_games = df[df['Home Team'] == team]
        # Away games for this team  
        away_games = df[df['Away Team'] == team]
        
        if len(home_games) > 0 and len(away_games) > 0:
            team_data = {'Team': team}
            
            # Take only the last 40 games for home and away
            home_games_last_40 = home_games.tail(40)
            away_games_last_40 = away_games.tail(40)
            
            # Alternate home and away for each stat (Home homeruns, Away homeruns, Home hits, Away hits, etc.)
            for stat_name in stat_names:
                home_col = f'Home {stat_name}'
                away_col = f'Away {stat_name}'
                
                team_data[f'Home {stat_name}'] = home_games_last_40[home_col].mean()
                team_data[f'Away {stat_name}'] = away_games_last_40[away_col].mean()
            
            team_avg_data.append(team_data)
    
    return pd.DataFrame(team_avg_data)

# Create the team averages dataframe
team_averages_df = create_team_averages(df)
team_averages_df = team_averages_df.sort_values('Team')
team_averages_df
```

```python
def get_player_era(player_name, home):
    player_id = mlb.get_people_id(player_name)[0]
    index = 0 if home else 1
    return float(mlb.get_player_stats(player_id, stats=['homeAndAway'], groups=['pitching'], season=2025)['pitching']['homeandaway'].splits[index].stat.era)
```

```python
def create_matchup_row(home_team, away_team, home_sp_name, away_sp_name, team_averages_df):
    """
    Creates a single row dataframe for a specific matchup with:
    - Home Team, Away Team
    - Alternating team stats (Home hr, Away hr, Home hits, Away hits, etc.)
    - Home SP name, Away SP name, Home SP ERA, Away SP ERA
    """
    # Get home team averages
    home_team_data = team_averages_df[team_averages_df['Team'] == home_team]
    away_team_data = team_averages_df[team_averages_df['Team'] == away_team]
    
    if home_team_data.empty or away_team_data.empty:
        print(f"Warning: Team data not found for {home_team} or {away_team}")
        return None
    
    # Create the matchup row
    matchup_data = {}
    
    # Add team names
    matchup_data['Home Team'] = home_team
    matchup_data['Away Team'] = away_team
    
    # Get stat columns (excluding Team column)
    stat_columns = [col for col in team_averages_df.columns if col != 'Team']
    
    # Extract unique stat names and alternate home/away
    home_stats = [col for col in stat_columns if col.startswith('Home ')]
    stat_names = [col.replace('Home ', '') for col in home_stats]
    
    # Add alternating stats
    for stat_name in stat_names:
        home_col = f'Home {stat_name}'
        away_col = f'Away {stat_name}'
        
        matchup_data[home_col] = home_team_data[home_col].iloc[0]
        matchup_data[away_col] = away_team_data[away_col].iloc[0]
    
    # Add starting pitcher information
    matchup_data['Home SP'] = home_sp_name
    matchup_data['Home SP ERA'] = get_player_era(home_sp_name, home=True)
    matchup_data['Away SP'] = away_sp_name
    matchup_data['Away SP ERA'] = get_player_era(away_sp_name, home=False)

    return pd.DataFrame([matchup_data])
```
